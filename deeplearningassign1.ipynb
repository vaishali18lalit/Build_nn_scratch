```python
# def logistic_regression(n): 
cost=[]
iter=5000
alpha=0.008
tolerance=0.1
mse=[]
no_features=3
theta= [random.randint(1,100) for _ in range(no_features+1)]
for i in range(0,iter,1):
    print(theta) 
    J=-1/2*(len(X_train))*np.sum(y_train*np.log(h(X_train))+ (1-y_train)*np.log(1-h(X_train)))
    cost.append(J)
    if i>0 and (np.all(np.abs(cost[-2]-cost[-1])<tolerance)):
            print("Convergence occured at {}th iteration ".format(i))
            break
    for j in range(no_features + 1):
            if j == 0:  # This is the bias term
                theta[j] = round(theta[j] - alpha * np.sum(h(X_train) - y_train), 4)
            else:
                theta[j] = round(theta[j] - alpha * np.sum((h(X_train) - y_train) * X_train.iloc[:,j-1]), 4)

    y_pred = h(X_test)
#     print("ji",len(y_pred),len(y_test))
    y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred]
    correct_pred=np.sum(y_pred_binary==y_test)
    total_pred=len(y_test)
    accuracy=correct_pred/total_pred

# #     mse.append(mean_absolute_error(y_test, y_pred))
#     accuracy.append(r2_score(y_test, y_pred))
#     y_pred1 = model.predict(X_test)
#     mse1.append(mean_absolute_error(y_test, y_pred1))
#     accuracy1.append(r2_score(y_test, y_pred1))
plt.figure(figsize=(15,8))
plt.subplot(5, 5, 1)
plt.plot(cost)
plt.title(f'Cost Function for Kfold')
plt.xlabel('Iterations')
#     plt.ylabel('Cost Function Value')
#     plt.subplot(5, 5, 2)
#     plt.plot(mse)
#     plt.title(f'Loss curve f')
#     plt.xlabel('Iterations')
#     plt.ylabel('Loss')
#     plt.subplot(5, 5, 3)
#     plt.plot(accuracy)
#     plt.title(f'accuracy curve for Kfold')
#     plt.xlabel('Iterations')
#     plt.ylabel('Accuracy')
#     plt.subplot(5, 5, 4)
print(accuracy)
```
